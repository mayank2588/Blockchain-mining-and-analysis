```python
import pandas as pd
import glob, os, json


meebits_json_dir = 'C:/Users/mayan/Downloads/SpringSemester 2022/Team Project/OpenSea-NFT-Stealer-master/images/meebits/image_data/'

json_pattern = os.path.join(meebits_json_dir, '*.json')
file_list = glob.glob(json_pattern)
```

```python
dfs = []
for file in file_list:
    with open(file) as f:
        json_data = pd.json_normalize(json.loads(f.read()))
        json_data['site'] = file.rsplit("/", 1)[-1]
    dfs.append(json_data)
df = pd.concat(dfs)
```

```python
display(df)
```

```python
df.to_csv(r'./meebits.csv', sep='\t', encoding='utf-8',header=True)
```
